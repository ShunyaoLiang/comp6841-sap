```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open ('2025-07-21-updated-benchmarks.json') as f:
    with_cache_miss, nothing = json.load(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Your two datasets
# with_cache_miss = [...]  # Your first dataset
# nothing = [...]          # Your second dataset

# Sort both datasets
sorted_cache_miss = np.sort(with_cache_miss)
sorted_nothing = np.sort(nothing)

# Create cumulative frequencies for both
cumulative_cache_miss = np.arange(1, len(sorted_cache_miss) + 1)
cumulative_nothing = np.arange(1, len(sorted_nothing) + 1)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot both datasets
plt.plot(sorted_cache_miss, cumulative_cache_miss, linewidth=2, 
         label=f'With Cache Miss ({len(with_cache_miss)} measurements)', color='red')
plt.plot(sorted_nothing, cumulative_nothing, linewidth=2, 
         label=f'Nothing ({len(nothing)} measurements)', color='blue')

plt.xscale('log')
plt.xlabel('Time (seconds) - log scale')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Distribution Comparison: With Cache Miss vs Nothing')

# Add reference lines for the larger dataset (or choose one specifically)
total_cache_miss = len(with_cache_miss)
total_nothing = len(nothing)

# Reference lines for cache miss dataset
plt.axhline(total_cache_miss * 0.5, color='red', linestyle=':', alpha=0.5, 
           label='50% Cache Miss data')
plt.axhline(total_cache_miss * 0.95, color='red', linestyle=':', alpha=0.5, 
           label='95% Cache Miss data')

# Reference lines for nothing dataset  
plt.axhline(total_nothing * 0.5, color='blue', linestyle=':', alpha=0.5, 
           label='50% Nothing data')
plt.axhline(total_nothing * 0.95, color='blue', linestyle=':', alpha=0.5, 
           label='95% Nothing data')

plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print some comparison statistics
print("Dataset Comparison:")
print(f"With Cache Miss - Min: {min(with_cache_miss):.6f}, Max: {max(with_cache_miss):.6f}, Mean: {np.mean(with_cache_miss):.6f}")
print(f"Nothing - Min: {min(nothing):.6f}, Max: {max(nothing):.6f}, Mean: {np.mean(nothing):.6f}")
print(f"Median difference: {np.median(with_cache_miss) - np.median(nothing):.6f}")
```
